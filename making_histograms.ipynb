```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_pickle('tickle_my.pkl')
df.groupby('genre').agg('count')
```

```python
df = pd.read_pickle('sad.pkl')
genres = ['Country', 'Electronic', 'Folk', 'Hip-Hop', 'Indie', 'Jazz', 'Metal',
         'Pop', 'R&B', 'Rock']
# genres = ['Country', 'Electronic', 'Hip-Hop', 'Indie', 'Jazz',
#          'R&B']
for name in genres:
    print(name)
    genre_name = name
    genre = df[genre_name]
    genre.columns = ['song_num', 'cos_sim']
    data = np.array(genre.tolist())
    data = data[np.logical_not(np.isnan(data))]
    print(len(data))
    binwidth = .05
    print(min(data))
    plt.gcf().clear()
    fig = plt.hist(data, histtype='bar', bins=np.arange(min(data), max(data) + binwidth, binwidth))
#     fig = plt.hist(data, histtype='bar', normed=True, bins=10)
    plt.title(genre_name+" Sadness Cosine Similarity")
    plt.xlabel("Cosine Similarity")
    plt.ylabel("Frequency")
    plt.savefig(genre_name+"_sad.png")
```

```python
dfh1 = pd.read_pickle('happytest.pkl')
dfh2 = pd.read_pickle('hatetest.pkl')
dfa = pd.read_pickle('lovetest.pkl')
dfs = pd.read_pickle('sadtest.pkl')
```

```python
import csv
test_songs = []
cos_sim = {}
with open('Lyrics.csv') as f:
    songs = csv.reader(f, delimiter=',')
    for row in songs:
        print(row[0])
        print(row[1])
        if row[0] != 'Artist':
            test_songs.append((row[0], row[1]))
```

```python
happy = dfh1.tolist()
i = 0
for value in happy:
    cos_sim[test_songs[i][1]] = []
    cos_sim[test_songs[i][1]].append(value)
    i+=1
hate = dfh2.tolist()
i = 0
for value in hate:
    cos_sim[test_songs[i][1]].append(value)
    i+=1
love = dfa.tolist()
i = 0
for value in love:
    cos_sim[test_songs[i][1]].append(value)
    i+=1
sad = dfs.tolist()
i = 0
for value in sad:
    cos_sim[test_songs[i][1]].append(value)
    i+=1
print(cos_sim)
```

```python
votes = {}

for artist in test_songs:
    votes[artist[1]] = [0, 0, 0, 0]
```

```python
with open('mturk_results.csv') as f:
    surveys = csv.reader(f, delimiter=',')
    for row in surveys:
        if row[0] != "HITId":
            responses = []
            for i in range(27,42):
                responses.append(row[i])
            j = 0
            for response in responses:
                if response == "Happy":
                    votes[test_songs[j][1]][0] += 1
                elif response == "Hate":
                    votes[test_songs[j][1]][1] += 1
                elif response == "Love":
                    votes[test_songs[j][1]][2] += 1
                elif response == "Sad":
                    votes[test_songs[j][1]][3] += 1
                j+=1
print(votes)            
```

```python
data = {}
data['Songs'] = []
for i in range(1,16):
    data['Songs'].append('Song '+str(i))
data['Happy'] = []
data['Hate'] = []
data['Love'] = []
data['Sad'] = []
for key in votes:
    data['Happy'].append(votes[key][0])
    data['Hate'].append(votes[key][1])
    data['Love'].append(votes[key][2])
    data['Sad'].append(votes[key][3])
    
df = pd.DataFrame(data, columns = ['Songs', 'Happy', 'Hate', 'Love', 'Sad'])

plt.gcf().clear()

# Setting the positions and width for the bars
pos = list(range(len(data['Songs']))) 
width = 0.2 

# Plotting the bars
fig, ax = plt.subplots(figsize=(15,5))

# Create a bar with pre_score data,
# in position pos,
plt.bar(pos, 
        #using df['pre_score'] data,
        df['Happy'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#EE3224', 
        # with label the first value in first_name
        label=df['Songs'][0]) 

# Create a bar with mid_score data,
# in position pos + some width buffer,
plt.bar([p + width for p in pos], 
        #using df['mid_score'] data,
        df['Hate'],
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#F78F1E', 
        # with label the second value in first_name
        label=df['Songs'][1]) 

# Create a bar with post_score data,
# in position pos + some width buffer,
plt.bar([p + width*2 for p in pos], 
        #using df['post_score'] data,
        df['Love'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#FFC222', 
        # with label the third value in first_name
        label=df['Songs'][2])

plt.bar([p + width*3 for p in pos], 
        #using df['post_score'] data,
        df['Sad'], 
        # of width
        width, 
        # with alpha 0.5
        alpha=0.5, 
        # with color
        color='#ffff00', 
        # with label the third value in first_name
        label=df['Songs'][3])

# Set the y axis label
ax.set_ylabel('Votes')

# Set the chart's title
ax.set_title('Mechanical Turk Results')

# Set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Set the labels for the x ticks
ax.set_xticklabels(df['Songs'])

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(df['Happy'] + df['Hate'] + df['Love'] + df['Sad'])] )
plt.legend(['Happy', 'Hate', 'Love', 'Sad'], loc='upper right')
# plt.gcf().clear()
plt.savefig("Mechanical Turk Results.png")
```
